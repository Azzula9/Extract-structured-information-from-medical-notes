<h2> Baseline model using regular expression</h2>

```python
# Function to transform a note to json
def note_to_json(note):
    data = {
        "patient_info": {
            "age": None,
            "gender": None
        },
        "visit_motivation": "",
        "symptoms": [],
        "vital_signs": {
            "heart_rate": {
                "value": None,
                "unit": "bpm"
            },
            "oxygen_saturation": {
                "value": None,
                "unit": "%"
            },
            "cholesterol_level": {
                "value": None,
                "unit": "mg/dL"
            },
            "glucose_level": {
                "value": None,
                "unit": "mg/dL"
            },
            "temperature": {
                "value": None,
                "unit": "°C"
            },
            "respiratory_rate": {
                "value": None,
                "unit": "breaths/min"
            },
            "blood_pressure": {
                "systolic": {
                    "value": None,
                    "unit": "mmHg"
                },
                "diastolic": {
                    "value": None,
                    "unit": "mmHg"
                }
            }
        }
    }

    # Extract age with multiple patterns
    age_patterns = [
        r'age[: ]+(\d+)',
        # 'age' matches the literal word "age".
        # '[: ]+' matches one or more occurrences of ':' or ' ' (space) after "age".
        # '(\d+)' captures one or more digits (representing the age).
        r'(\d+)[-\s]?year[-\s]?old',
        # '[-\s]?' optionally(? means optionally) matches a hyphen (-) or whitespace (\s).
        # 'year' matches the literal word "year".
        # 'old' matches the literal word "old".
        r'age is (\d+)',
        r'(\d+)\s*yo\b'
        # '\s*' matches zero or more whitespace characters.
        # 'yo' matches the literal "yo"(refers to years old).
        # '\b' asserts a word boundary, ensuring "yo" is a complete word.
    ]
    for pattern in age_patterns:
        match = re.search(pattern, note, re.IGNORECASE)
        if match:
            data['patient_info']['age'] = int(match.group(1))
            break

    # Extract gender with multiple patterns
    gender_patterns = [
        r'gender[: ]+(male|female)',
        # '[: ]+' matches one or more ':' or ' ' (space)characters.
        # '(male|female)' captures either "male" or "female"(| means either).
        r'sex[: ]+(male|female)',
        r'\b(male|female)\b patient'
        # '\b' asserts a word boundary before and after "male" or "female".
        r'gender is (male|female)'
    ]
    for pattern in gender_patterns:
        match = re.search(pattern, note, re.IGNORECASE)
        if match:
            data['patient_info']['gender'] = match.group(1).capitalize()
            break

    # Extract visit motivation
    motivation_patterns = [
        r'visit motivation[:\- ]+([^\n]+)',
        # '[:\- ]+' matches one or more ':', '-', or ' ' (space).
        # '([^\n]+)' captures everything until a newline ('\n'), representing the motivation.
        r'reason for visit[:\- ]+([^\n]+)',
        r'chief complaint[:\- ]+([^\n]+)',
        r'presenting with[:\- ]+([^\n]+)'
    ]
    for pattern in motivation_patterns:
        match = re.search(pattern, note, re.IGNORECASE)
        if match:
            data['visit_motivation'] = match.group(1).strip()
            break

    # Extract symptoms
    symptoms_patterns = [
        r'symptoms?[:\- ]+([^\n]+)',
        # 'symptoms?' matches the word "symptom" or "symptoms" (optional 's').
        # '[:\- ]+' matches one or more ':', '-', or ' '.
        # '([^\n]+)' captures everything until a newline.
        r'complaints?[:\- ]+([^\n]+)',
        r'patient reports[:\- ]+([^\n]+)',
        r'symptoms include[:\- ]+([^\n]+)',
        r'presenting with[:\- ]+([^\n]+)'
    ]
    for pattern in symptoms_patterns:
        match = re.search(pattern, note, re.IGNORECASE)
        if match:
            symptoms_list = re.split(r',|\band\b', match.group(1))
            data['symptoms'] = [symptom.strip() for symptom in symptoms_list if symptom.strip()]
            break

    # Extract vital signs
    vital_patterns = {
        'heart_rate': [r'heart rate[:\- ]+(\d+)', r'H[ea]*R[:\- ]+(\d+)'],
        # '[:\- ]+' matches one or more ':', '-', or ' '.
        # '(\d+)' captures one or more digits.

        #r'H[ea]*R[:\- ]+(\d+)'  # Matches "HR: 75"
        # 'H[ea]*R' matches "HR", "HeR", "HaR", etc.
        # '[:\- ]+' matches one or more ':', '-', or ' '.
        # '(\d+)' captures one or more digits.
        'oxygen_saturation': [r'oxygen saturation[:\- ]+(\d+)', r'SpO2[:\- ]+(\d+)%?'],
        # r'SpO2[:\- ]+(\d+)%?'  # Matches "SpO2: 95" or "SpO2: 95%"
        # 'SpO2' matches the literal phrase "SpO2".
        # '[:\- ]+' matches one or more ':', '-', or ' '.
        # '(\d+)' captures one or more digits.
        # '%?' optionally matches the '%' character.
        'cholesterol_level': [r'cholesterol level[:\- ]+(\d+)', r'cholesterol[:\- ]+(\d+)'],
        'glucose_level': [r'glucose level[:\- ]+(\d+)', r'blood glucose[:\- ]+(\d+)'],
        'temperature': [r'temperature[:\- ]+([\d\.]+)', r'temp[:\- ]+([\d\.]+)'],
        'respiratory_rate': [r'respiratory rate[:\- ]+(\d+)', r'respirations[:\- ]+(\d+)'],
        'blood_pressure': [r'blood pressure[:\- ]+(\d+)/(\d+)', r'BP[:\- ]+(\d+)/(\d+)']
    }

    for key, patterns_list in vital_patterns.items():
        for pattern in patterns_list:
            match = re.search(pattern, note, re.IGNORECASE)
            if match:
                if key == 'blood_pressure':
                    data['vital_signs']['blood_pressure']['systolic']['value'] = float(match.group(1))
                    data['vital_signs']['blood_pressure']['diastolic']['value'] = float(match.group(2))
                else:
                    data['vital_signs'][key]['value'] = float(match.group(1))
                break

    return json.dumps(data)

# Apply transformation to test data
test['predictions'] = test['Note'].apply(note_to_json)

# Create submission file
# submission_df = test[['ID', 'json']]
# submission_df.to_csv('submission.csv', index=False)
```

This Python function calculates the accuracy of all fields between two JSON structures (predictions and ground_truths). The accuracy is based on how many fields in the predicted JSON exactly match the corresponding fields in the ground truth JSON.

```python
import json

def calculate_all_fields_accuracy(predictions, ground_truths):
    total_fields = 0
    correct_fields = 0

    for pred, gt in zip(predictions, ground_truths):
        pred_dict = json.loads(pred)
        gt_dict = json.loads(gt)
        
        # Recursive comparison of all fields in the JSON
        def compare_dicts(d1, d2):
            nonlocal total_fields, correct_fields
            for key in d1:
                total_fields += 1
                if key in d2:
                    if isinstance(d1[key], dict) and isinstance(d2[key], dict):
                        compare_dicts(d1[key], d2[key])
                    elif d1[key] == d2[key]:
                        correct_fields += 1

        compare_dicts(pred_dict, gt_dict)

    return correct_fields / total_fields if total_fields > 0 else 0
```

```python
def calculate_exact_match_accuracy(predictions, ground_truths):
    exact_matches = sum(1 for pred, gt in zip(predictions, ground_truths) if pred == gt)
    return exact_matches / len(predictions) if predictions else 0
```

```python
# Assuming the test dataset has a ground truth column `ground_truth_json`
predictions = test['predictions'].tolist()
ground_truths = test['json'].tolist()

# Calculate accuracies
field_accuracy = calculate_all_fields_accuracy(predictions, ground_truths)
exact_match_accuracy = calculate_exact_match_accuracy(predictions, ground_truths)

print(f"Field-Level Accuracy (All Fields): {field_accuracy:.2%}")
print(f"Exact Match Accuracy: {exact_match_accuracy:.2%}")
```
